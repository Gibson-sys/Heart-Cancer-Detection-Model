## Dealing with Outliers

```python
# Removing outliers from each features

Q1,Q3 = np.percentile(Heart_df["chol"],[25,75])
IQR = Q3 - Q1 
lower_fence = Q1 - (1.5 * IQR)
upper_fence = Q3 + (1.5 * IQR)
Heart_df[(Heart_df["chol"] < lower_fence) | (Heart_df["chol"] > upper_fence)]
bf = Heart_df[(Heart_df["chol"] >= lower_fence) & (Heart_df["chol"] <= upper_fence)]
bf = pd.DataFrame(bf)

q1,q3 = np.percentile(bf["trtbps"],[25,75])
iqr = q3 - q1 
lower_fence = q1 - (1.5 * iqr)
upper_fence = q3 + (1.5 * iqr)
bf[(bf["trtbps"] < lower_fence) | (bf["trtbps"] > upper_fence)]
df = bf[(bf["trtbps"] >= lower_fence) & (bf["trtbps"] <= upper_fence)]

qt1,qt3 = np.percentile(df["fbs"],[25,75])
inqr = qt3 - qt1 
lower_fence = qt1 - (1.5 * inqr)
upper_fence = qt3 + (1.5 * inqr)
df[(df["fbs"] < lower_fence) | (df["fbs"] > upper_fence)]
nf = df[(df["fbs"] >= lower_fence) & (df["fbs"] <= upper_fence)]

qtl1,qtl3 = np.percentile(nf["thalachh"],[25,75])
inqrt = qtl3 - qtl1 
lower_fence = qtl1 - (1.5 * inqrt)
upper_fence = qtl3 + (1.5 * inqrt)
nf[(nf["thalachh"] < lower_fence) | (nf["thalachh"] > upper_fence)]
mf = nf[(nf["thalachh"] >= lower_fence) & (nf["thalachh"] <= upper_fence)]

qtl1_,qtl3_ = np.percentile(mf["oldpeak"],[25,75])
inqrt_ = qtl3_ - qtl1_ 
lower_fence = qtl1_ - (1.5 * inqrt_)
upper_fence = qtl3_ + (1.5 * inqrt_)
mf[(mf["oldpeak"] < lower_fence) | (mf["oldpeak"] > upper_fence)]
gf = mf[(mf["oldpeak"] >= lower_fence) & (mf["oldpeak"] <= upper_fence)]

m,n = np.percentile(gf["caa"],[25,75])
u = n - m 
lower_fence = m - (1.5 * u)
upper_fence = n + (1.5 * u)
gf[(gf["caa"] < lower_fence) | (gf["caa"] > upper_fence)]
ff = gf[(gf["caa"] >= lower_fence) & (gf["caa"] <= upper_fence)]

x,y = np.percentile(ff["thall"],[25,75])
z = y - x 
lower_fence = x - (1.5 * z)
upper_fence = y + (1.5 * z)
ff[(ff["thall"] < lower_fence) | (ff["thall"] > upper_fence)]
af = ff[(ff["thall"] >= lower_fence) & (ff["thall"] <= upper_fence)]

QU1,QU3 = np.percentile(af["chol"],[25,75])
IQUR= QU3 - QU1 
lower_fence = QU1- (1.5 * IQUR)
upper_fence = QU3 + (1.5 * IQUR)
af[(af["chol"] < lower_fence) | (af["chol"] > upper_fence)]
real_df = af[(af["chol"] >= lower_fence) & (af["chol"] <= upper_fence)]

i,h= np.percentile(real_df["thalachh"],[25,75])
gt = h - i
lower_fence = i- (1.5 * gt)
upper_fence = h + (1.5 * gt)
real_df[(real_df["thalachh"] < lower_fence) | (real_df["thalachh"] > upper_fence)]
finish_df= real_df[(real_df["thalachh"] >= lower_fence) & (real_df["thalachh"] <= upper_fence)]
```

## Exploring Custom Ensemble Learning Technique

```python
# importing models
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.pipeline import Pipeline
from sklearn.ensemble import VotingClassifier
```

```python
#list for the models
models = []

# creating a pipeline for each model
randomizer_pipeline = Pipeline([("x",RandomForestClassifier())])
models.append(("random",randomizer_pipeline))


logistic_pipeline = Pipeline([("x",LogisticRegression(max_iter=10000))])
models.append(("logistic",logistic_pipeline))


svm_pipeline = Pipeline([("x", SVC(max_iter= 1000))])
models.append(("support",svm_pipeline))


tree_pipeline = Pipeline([("x",DecisionTreeClassifier())])
models.append(("tree",tree_pipeline))


gradient_pipeline = Pipeline([("x", GradientBoostingClassifier())])
models.append(("gradient",gradient_pipeline))

# defining voting classifier
voting_classifier = VotingClassifier(models,voting = "hard")
voting_classifier
```

```python
models
```

```python
cv = RepeatedStratifiedKFold(n_splits = 10, n_repeats = 3, random_state = 1)

#scoring
cross_val = cross_val_score(voting_classifier,X,Y,cv = cv,scoring="accuracy")

print(cross_val)
print("-----------------------------------------")

print(cross_val.mean())
print("--------------------------------------")

#predict with scoring
cross_predict = cross_val_predict(voting_classifier,X,Y,cv = 5)
print(accuracy_score(Y,cross_predict))
```
